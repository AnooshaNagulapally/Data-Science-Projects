```python
# Import libraries BeautifulSoup and Pandas
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
# get the url from  2cheapcars website & assign  to variable url 
url = "https://www.2cheapcars.co.nz/used-vehicles"
```

```python
# read the url from response variable
response = requests.get(url)
response.status_code
```

```python
# to read the content on the page 
soup = BeautifulSoup(response.content, 'html.parser')
```

```python
# get  the vehicle name and count the vehicle names in one page
lists = soup.find_all('li', class_="vehicle")
len(lists)
```

```python
# get the  car modelname and year   
lists[0].find('h6').get_text().replace('\r\n', '')
```

```python
# get the kilometer , version and cc
lists[0].find('ul',{'class':'vehicle-specs hide-for-small-only'}).get_text().split('\n')
```

```python
# get the car price
lists[0].find('div',{'class':'price'}).get_text().replace('\r\n', '')
```

```python
# get the branch name
lists[0].find('div',{'class':'branch'}).get_text()
```

```python
#  using for loop to get the data for whole page 
model = []
information = []
price = []
branch= []


for list in lists:
    try:
        model.append(list.find('h6').get_text().replace('\r\n', ''))
    except:
        model.append('n/a')
        
    try:
        information.append(list.find('ul',{'class':'vehicle-specs hide-for-small-only'}).get_text().split('\n'))
    except:
        information.append('n/a')
        
    try:
        price.append(list.find('div',{'class':'price'}).get_text().replace('\r\n', ''))
    except:
        price.append('n/a')
        
        
    try:
        branch.append(list.find('div',{'class':'branch'}).get_text())
    except:
        branch.append('n/a')
    
```

```python
# Create data frame with Model, Information , Price &  Branch column names 
car_data = pd.DataFrame({'Model':model, 'Information':information, 'Price':price, 'Branch':branch })
```

```python
car_data
```

```python
# pagination to get data for all pages 
model = []
information = []
price = []
branch = []

for i in range(1, 17):
    url = "https://www.2cheapcars.co.nz/used-vehicles?page=' + str(i) +' "
    response = requests.get(url)
    soup = BeautifulSoup(response.content, 'html.parser')
    lists = soup.find_all('li', class_="vehicle")
    
    for list in lists:
        try:
            model.append(list.find('h6').get_text().replace('\r\n', ''))
        except:
            model.append('n/a')
        
        try:
            information.append(list.find('ul',{'class':'vehicle-specs hide-for-small-only'}).get_text().split('\n'))
        except:
            information.append('n/a')
        
        try:
            price.append(list.find('div',{'class':'price'}).get_text().replace('\r\n', ''))
        except:
            price.append('n/a')
        
        
        try:
            branch.append(list.find('div',{'class':'branch'}).get_text())
        except:
            branch.append('n/a')
        
    
```

```python
# create data frame with columns and values
car_data = pd.DataFrame({'Model':model, 'Information':information, 'Price':price, 'Branch':branch })
car_data
```

```python
# data Processing , create one column in to list 
car_data['Information']
split_information = pd.DataFrame(car_data['Information'].tolist())
split_information
```

```python
# splitting data and rename the column names
split_1 = split_information.rename(columns={0:"extr" ,1:"km",2:"version",3:"type",4:"cc" ,5:"extra"})
car_data = pd.concat([car_data, split_1], axis=1)
car_data
```

```python
# drop some unnecessary columns
car_data = car_data.drop('extr', axis=1)
car_data = car_data.drop('extra', axis=1)
```

```python
# drop some columns
car_data = car_data.drop('Information', axis=1)
car_data
```

```python
# replace some string values
car_data['km'] = car_data['km'].str.replace('km','')
car_data['cc'] = car_data['cc'].str.replace('cc','')
car_data['Price'] = car_data['Price'].str.replace('\n','')
```

```python
# drop the columns 
split_3 = split_3.drop('e' , axis = 1)
split_3 = split_3.drop('ex' , axis = 1)
split_3 = split_3.drop('ext' , axis = 1)
split_3
```
