```julia
# Scraping turners.co.nz cars data in julia 
# getting web page
Url = "https://www.turners.co.nz/Cars/Used-Cars-for-Sale/?sortorder=7&pagesize=24&pageno=1"
cars_page = HTTP.get(Url)
```

```julia
# parse the html 
parsed_page = cars_page.body |> String |> parsehtml
```

```julia
#getting the car names, location, price data using css selector
car_name = Selector(".search-header")
Carname_title = eachmatch(car_name, parsed_page.root)
location = Selector(".location .value")
Location = eachmatch(location,parsed_page.root)
price = Selector(".buyNowDiscountPrice")
Price = eachmatch(price, parsed_page.root)
finance= Selector(".finance span")
Finance= eachmatch(finance, parsed_page.root)
odometer= Selector(".odometer span")
Odometer = eachmatch(odometer, parsed_page.root)
```

```julia
# getting text 
result_string = nodeText.(Carname_title) 
join(result_string, " ") |> println 
result_string1 = nodeText.(Location) 
join(result_string1, " ") |> println 
result_string2 = nodeText.(Price) 
join(result_string2, " ") |> println 
result_string3 = nodeText.(Finance) 
join(result_string3, " ") |> println 
result_string4 = nodeText.(Odometer) 
join(result_string4, " ") |> println 
```

```julia
# using function  to make dataframe
function get_cars_info(link)
  response = HTTP.request("GET","https://www.turners.co.nz/Cars/Used-Cars-for-Sale/?sortorder=7&pagesize=24&pageno=1") 
  parsed_page = response.body |> String |> parsehtml
  parsed_page_part = eachmatch(Carname_title, Location, Price, Finance, Odometer, parsed_page.root)
  array_string = nodeText.(result_string, result_string1, result_string2, result_string3, result_string4, parsed_page_part) 
  result_str = join(array_string, " ")
  df = DataFrame(Link = link, Info = result_str)
  return df
end
```
